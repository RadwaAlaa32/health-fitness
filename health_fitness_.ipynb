```python
target = 'fitness_level'
features = [col for col in df.columns if col != target]
```

```python
features = [col for col in df.columns if col != target and df[col].dtype != 'object']
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_features, df[target], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

```python
coefficients = pd.DataFrame({
    'Feature': features,  # Use the features list used for training the model
    'Coefficient': model.coef_
}).sort_values(by='Coefficient', ascending=False)

print(coefficients)
```

```python
# Visualize the results (Corrected)
plt.scatter(y_test, y_pred, color="blue", label="Actual vs. Predicted") # Changed X_test to y_test
plt.xlabel("Actual Values (Fitness Level)")  # Changed label
plt.ylabel("Predicted Values (Fitness Level)") # Changed label
plt.title("Linear Regression Results")
plt.legend()
plt.show()
```
